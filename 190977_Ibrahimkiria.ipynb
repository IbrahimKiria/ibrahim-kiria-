# Removal of Missing Values : Before and After

```python
from sklearn.impute import SimpleImputer

print("Missing values before imputation:\n", life_expectancy_data.isnull().sum())

median_imputer = SimpleImputer(strategy='median')

columns_with_missing = life_expectancy_data.columns[life_expectancy_data.isnull().any()]

life_expectancy_data[columns_with_missing] = median_imputer.fit_transform(life_expectancy_data[columns_with_missing])

missing_values_after = life_expectancy_data.isnull().sum()
print("Missing values after imputation:\n", missing_values_after)

updated_file_path = 'life_expectancy_data_updated.csv'
life_expectancy_data.to_csv(updated_file_path, index=False)

print("Missing values treated and updated file saved.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load the dataset
updated_file_path = 'life_expectancy_data_updated.csv'
life_expectancy_data = pd.read_csv(updated_file_path)

# Handle missing values (using median imputation)
life_expectancy_data.fillna(life_expectancy_data.median(numeric_only=True), inplace=True)

# Feature Engineering
# Encoding categorical features
categorical_features = ['Region', 'IncomeGroup']
life_expectancy_data = pd.get_dummies(life_expectancy_data, columns=categorical_features, drop_first=True)

# Preparing the features and target
X = life_expectancy_data.drop(columns=['Life Expectancy World Bank'])
y = life_expectancy_data['Life Expectancy World Bank']

# Feature Scaling 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.select_dtypes(include=['float64', 'int64']))

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Plot true values vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, label='Random Forest Predictions', color='green')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs Predictions')
plt.legend()
plt.grid(True)
plt.show()

```
